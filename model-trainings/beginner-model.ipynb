```python
from ultralytics import YOLO
import supervision as sv
import numpy as np
import cv2
import datetime
```

# Training

```python
PRETRAINED_MODEL = 'yolov8n-cls.pt'
DATA_PATH = '/content/drive/MyDrive/courses/fall2023/Internet of Things/Final Project/Image Classification Training YOLOv8/dataset'
EPOCH = 5
```

```python
model = YOLO(PRETRAINED_MODEL) # load a pretrained model

model.train(data=DATA_PATH, epochs=EPOCH) # train the model
```

# Inference -- with a recorded video

```python
VIDEO_PATH = "/content/drive/MyDrive/courses/fall2023/Internet of Things/Final Project/Image Classification Training YOLOv8/VideoToTest/1 Hour BEGINNER Yoga for Strength, Balance & Flexibility - NO PROPS.mp4"
BEST_MODEL = "/content/drive/MyDrive/courses/fall2023/Internet of Things/Final Project/models/Jan-07/classify/train/weights/best.pt"

CLASSES = [
    "downdog",
    "goddess",
    "plank",
    "tree",
    "warrior2",
]

THRESH = 0.5

DEBUG = True
```

```python
video = cv2.VideoCapture(VIDEO_PATH) # get the video

model = YOLO(BEST_MODEL) # get the model

predictions = []


while video.isOpened():

    _, frame = video.read()
    results = model(frame, verbose=False, conf=THRESH)

    for r in results:

        cls = CLASSES[r.probs.top1]
        score = r.probs.top1conf.cpu()

        if not predictions or predictions[-1]["class"] != cls:

            if predictions:
                predictions[-1]["final_timestamp"] = video.get(cv2.CAP_PROP_POS_MSEC)
                predictions[-1]["duration"] = datetime.timedelta(milliseconds=predictions[-1]["final_timestamp"] - predictions[-1]["start_timestamp"])
                predictions[-1]["avg_score"] = np.average(predictions[-1]["scores"])

            predictions.append(
                {
                    "class": cls,
                    "scores": [score],
                    "avg_score": None,
                    "final_timestamp": None,
                    "start_timestamp": video.get(cv2.CAP_PROP_POS_MSEC),
                    "duration": None
                }
            )

        else:
            predictions[-1]["scores"].append(score)

        if DEBUG:
            print("*"*20)
            print(f"Pose: {cls}, Score: {score:.2f}")

video.release()
cv2.destroyAllWindows()
```

```python
for pred in predictions:
    print(f"Pose: {pred['class']} \t Score: {pred['avg_score']} \t Duration: {pred['duration']}")
```
